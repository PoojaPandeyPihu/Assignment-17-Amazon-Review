```python
from bs4 import BeautifulSoup as bs
import requests
```

```python
link = 'https://www.amazon.in/Noise-ColorFit-Smartwatch-Display-Monitors/product-reviews/B097R3CN3Y/ref=cm_cr_dp_d_show_all_btm?ie=UTF8&reviewerType=all_reviews'
```

```python
page = requests.get(link)
```

```python
page
```

```python
page.content
```

```python
soup = bs(page.content,'html.parser')
```

```python
print(soup.prettify())
```

```python
name = soup.find_all('span',class_ = 'a-profile-name')
```

```python
name
```

```python
len(name)
```

```python
cust_name = []
for i in range(0,len(name)):
  cust_name.append(name[i].get_text())
cust_name
```

```python
cust_name.pop(0)
```

```python
cust_name.pop(0)
```

```python
cust_name
```

```python
title = soup.find_all('a',class_='review-title-content')
```

```python
title
```

```python
review_title = []
for i in range(0,len(title)):
  review_title.append(title[i].get_text())
review_title
```

```python
review_title[:] =[titles.lstrip('\n') for titles in review_title]
```

```python
review_title
```

```python
review_title[:] =[titles.rstrip('\n') for titles in review_title]
review_title
```

```python
rating = soup.find_all('i',class_= 'review-rating')
rating
```

```python
rate = []
for i in range(0,len(rating)):
  rate.append(rating[i].get_text())
rate
```

```python
rate.pop(0)
```

```python
rate.pop(0)
```

```python
rate
```

```python
review = soup.find_all("span",{"data-hook":"review-body"})
review
```

```python
review_content = []
for i in range(0,len(review)):
    review_content.append(review[i].get_text())
review_content
```

```python
review_content[:] = [reviews.lstrip('\n') for reviews in review_content]
review_content
```

```python
review_content[:] = [reviews.lstrip('\r') for reviews in review_content]
review_content
```

```python
len(review_content)
```

```python
cust_name
review_title
rate
review_content
```

```python
import pandas as pd
```

```python
df=pd.DataFrame()
```

```python
df
```

```python
df['Customer Name'] = cust_name
df
```

```python
df['Review Title'] = review_title
df['Ratings'] = rate
df['Reviews'] = review_content


```
